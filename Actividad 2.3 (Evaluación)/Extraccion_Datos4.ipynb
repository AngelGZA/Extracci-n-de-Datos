```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

CARGA DEL ARCHIVO DEL PAIS .XLSX

```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_csv('Datos_Los_Angeles_Estados_Unidos.csv')
df
```

Filtro por 5 host_name

```python
#Filtro 1 usando el filtro por  objetos.
filtro1=df[df["host_name"].isin(["Ida", "Jeremy", "Jonathan", "Raven", "Yasmin"])]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable1_Evaluacion.csv")
```

Filtros de los host que se unieron a Airbnb "antes del año 2019"

```python
#Filtro 2 usando filtro por comparacion y usando un datetime dado que habia problemas con la fecha.
df["host_since"] = pd.to_datetime(df["host_since"], errors='coerce')
filtro2=df[df["host_since"] < "01-01-2019"]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Entregable2_Evaluacion.csv")
```

C. Los registros de los host que responden "a mas tardar en 1 dia"

```python
#filtro por comparacion
filtro3=df[df["host_response_time"]== "within a day"]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Entregable3_Evaluacion.csv")
```

D. Los registros que tengan "al menos 5 accommodates"

```python
#filtro por comparacion
filtro4=df[df["accommodates"]<= 5]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Entregable4_Evaluacion.csv")
```

E. Los registros de los tipos de cuarto: "Entire Home" y "Private room"

```python
filtro5=df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("Entregable5_Evaluacion.csv")
```

F. Los registros que "a lo mucho cuenten con 4 camas"

```python
#filtro por comparacion
filtro6=df[df["beds"] <= 4]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Entregable6_Evaluacion.csv")
```

G. Los registros que "tengan 1 baño o que sean superhost" pero con la condicion de que "todos sean instant_bookable"

```python
#filtro o
filtro7=df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"]== "t")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Entregable7_Evaluacion.csv")
```

H. Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3"

```python
#filtro Y
filtro8=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Entregable8_Evaluacion.csv")
```
